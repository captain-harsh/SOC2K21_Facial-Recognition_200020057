```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../Graduate Admissions.csv")
df.columns=[x.strip() for x in df.columns]
```

```python
df.head()
```

```python
sns.pairplot(df.drop('Serial No.',axis=1))
```

```python
def boxplot(y_column):
    columns=['University Rating','SOP','LOR','Research']
    fig = plt.figure(figsize=(30, 5))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    for i in range(1,len(columns)+1):
        ax=fig.add_subplot(1, len(columns), i)
        ax=sns.boxplot(x=columns[i-1],y=y_column,data=df)
boxplot('Chance of Admit')
```

```python
plt.rcParams['figure.figsize'] = (18, 9)
plt.style.use('ggplot')

sns.violinplot(df['University Rating'], df['Chance of Admit'], palette = 'copper')
plt.title('University Ratings vs Chance of Admission', fontsize = 20)
plt.show()
```

```python
df_sort = df.sort_values(by=df.columns[-1],ascending=False)
df_sort
```
